```python
# Mengimport library dan membaca data
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DM_Tabiin_DE02/Transaction_sales.csv')
df
```

```python
# Membuat kolom baru yang berisi semua produk yang dibeli dalam satu transaksi
df['All Products'] = df[['Product 1', 'Product 2', 'Product 3', 'Product 4']].agg(lambda x: ', '.join(x.dropna()), axis=1)
print(df[['Transaction ID', 'All Products']])
```

```python
# Menampilkan data asli produk
df.values
```

```python
# Transformasi data ke bentuk array
data = list(df['All Products'].apply(lambda x: x.split(",")))
data
```

```python
# Membuat algoritma Apriori
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

data_encoded = te.fit(data).transform(data)
df = pd.DataFrame(data_encoded, columns = te.columns_)
df = df.astype(int)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support = 0.02, use_colnames = True)

rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1.0)
```
